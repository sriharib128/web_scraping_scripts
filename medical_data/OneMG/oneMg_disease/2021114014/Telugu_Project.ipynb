```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint

url = 'https://www.sakshi.com/telugu-news/national/uttar-pradesh-dgp-mukul-goel-removed-1455460'
req = requests.get(url)
print(req.content)
```

```python
import re
html_content = req.content
soup = BeautifulSoup(html_content, 'html5lib')
from urllib.parse import urljoin 
news_corpus = soup.find_all('p')
corpus = ""
count = 0
for i in news_corpus:
  corpus=corpus+re.sub(r"[a-zA-Z0-9 |,?:!]"," ",i.text)
  
for j in range(len(corpus)):
     if corpus[j] == '.':
      count = count + 1
print(corpus)
print(count)
```

```python
#LEMMATIZATION

filename = 'Telugu.txt'
file = open(filename,'r')

Lemma_text = []

with open ("Telugu_Lemma.txt", 'w') as out:
    pipe = stanza.Pipeline('te', preocessors='tokensize,lemma',use_gpu = False)
    for line in file:
        sen = pipe(line)
        for each_line in sen.sentences:
            for word in each_line.words:
                Lemma_text.append(word.lemma)
                print(word.text, " ---> ", word.lemma, file= out)
file.close()

print(Lemma_text)


```

```python

from collections import Counter

Counter(Lemma_text)
count_check = Counter(Lemma_text)
print(count_check.most_common())
my_dict = dict(count_check.most_common(100))
pprint.pprint(my_dict)
```

```python

from matplotlib import pyplot as plt
import numpy as np
courses = list(my_dict.keys())
frequencies = list(my_dict.values())

fig = plt.figure(figsize=(10,5))
plt.xlabel("Observed words")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.title("Word Distribution in Telugu")
plt.bar(courses, frequencies, width=0.2)
plt.plot(courses,frequencies)
plt.show()
```

```python
#POS TAGGING

filename = 'Telugu.txt'
file = open(filename,'r')

with open ("telugu_pos.txt", 'w') as out:
    pipe = stanza.Pipeline('te', preocessors='tokensize,lemma,pos',use_gpu = False)
    for line in file:
        sen = pipe(line)
        for each_line in sen.sentences:
            for word in each_line.words:
                print("word: ",word.text, "POS: ", word.pos, file= out)
file.close()


```
